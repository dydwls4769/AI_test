### 이미지 자르기

- 이미지 잘라서 새로운 창에 표시

```python
import cv2
img = cv2.imread('img.jpg')
# img.shape (427, 640, 3)
crop = img[81:351 , 31:285] #세로 기준값 , 가로기준값

cv2.imshow('img',img)
cv2.imshow('crop',crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

이미지 대칭

### 이미지 회전

- 시계 방향 90도 회전

```python
import cv2
img = cv2.imread('img.jpg')

rotate90 =  cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) #시계방향 90도 회전
rotate180 =  cv2.rotate(img, cv2.ROTATE_180) #180도 회전
rotate270 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE) #시계반대방향 90도회전


cv2.imshow('img',img)
cv2.imshow('rotate90',rotate90)
cv2.imshow('rotate180',rotate180)
cv2.imshow('rotate270',rotate270)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 이미지 블러(흐리게)

- 커널 사이즈 변화에 따른 블러(가우시안 블러)

```python
img= cv2.imread('img.jpg')

#커널 사이즈 => (3,3) (5,5) (7,7)
kenel_33=cv2.GaussianBlur(img, (3,3), 0)
kenel_55=cv2.GaussianBlur(img, (5,5), 0)
kenel_77=cv2.GaussianBlur(img, (7,7), 0)

cv2.imshow('img',img)
cv2.imshow('kenel_33',kenel_33)
cv2.imshow('kenel_55',kenel_55)
cv2.imshow('kenel_77',kenel_77)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 표준편차에 따른 블러

```python
img= cv2.imread('img.jpg')

#커널 사이즈 => (3,3) (5,5) (7,7)
sigma_1=cv2.GaussianBlur(img, (0,0), 1) #sigmaX 가우시안 블러의 x축 방향의 표준편차 
sigma_2=cv2.GaussianBlur(img, (0,0), 2)
sigma_3=cv2.GaussianBlur(img, (0,0), 3)

cv2.imshow('img',img)
cv2.imshow('sigma_1',sigma_1)
cv2.imshow('sigma_2',sigma_2)
cv2.imshow('sigma_3',sigma_3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 이미지 다림질 효과 (원근감)

```python
import numpy as np
img= cv2.imread('news.jpg',cv2.IMREAD_GRAYSCALE)

width , height = 640,200 #너비 높이

src = np.array([[508,381],[1020,371],[1120,587],[457,600]], dtype = np.float32)
dst = np.array([[0,0],[width,0],[width,height],[0,height]], dtype = np.float32)
#4개 꼭지점은 시계방향으로 좌표값 입력

matrix = cv2.getPerspectiveTransform(src, dst)
res= cv2.warpPerspective(img, matrix, (width , height))

cv2.imshow('img',img)
cv2.imshow('res',res)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 이미지를 다운받고 이름만 집어넣었을때 4개의 점을 찍으면 평평하게 보일수 있도록 만들기

```python
src = []
img = cv2.imread('card.jpg', cv2.IMREAD_GRAYSCALE)
COLOR = (255, 0, 0)

def mouse_handler(event, x, y, flag, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        src.append((x, y))
        if len(src) == 4:
            show_result()

        for point in src:
            cv2.circle(img, point, 10, COLOR, cv2.FILLED)
        cv2.imshow('img', img)

def show_result():
    pts = np.array(src, dtype=np.float32)                        
    width = int(max(#(두번째 값의 x값 - 첫번째 값의 x값)**2 + (두번째 값의 y값 - 첫번째 값의 y값 , (세번째 값의 x값 - 4번째 값의 x값)**2 +
        np.sqrt((src[1][0]-src[0][0])**2 + (src[1][1]-src[0][1])**2),
        np.sqrt((src[2][0]-src[3][0])**2 + (src[2][1]-src[3][1])**2)
    ))
    height = int(max(
        np.sqrt((src[3][0]-src[0][0])**2 + (src[3][1]-src[0][1])**2),
        np.sqrt((src[2][0]-src[1][0])**2 + (src[2][1]-src[1][1])**2)
    ))
    dst = np.array([[0,0],[width,0],[width,height],[0,height]], dtype=np.float32)
    matrix = cv2.getPerspectiveTransform(np.array(src, dtype=np.float32), dst)
    res = cv2.warpPerspective(img, matrix, (width, height))
    cv2.imshow('img', img)
    cv2.imshow('res', res)

cv2.namedWindow('img')
cv2.setMouseCallback('img', mouse_handler)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
