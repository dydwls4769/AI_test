```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# 봇 감지 우회 옵션
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")

# 1. 구글 → yes24 파이썬 종합베스트 이동
driver.get("https://www.google.com")
time.sleep(2)

search_box = driver.find_element(By.ID, "APjFqb")
search_box.click()
search_box.send_keys("yes24 파이썬 종합베스트")
time.sleep(1)
search_box.send_keys(Keys.ENTER)
time.sleep(3)

results = driver.find_elements(By.CSS_SELECTOR, "h3.LC20lb")
for result in results:
    if "파이썬" in result.text and "베스트" in result.text:
        print(f"클릭: {result.text}")
        result.click()
        break

time.sleep(3)
print(f"이동된 URL: {driver.current_url}\n" + "="*60)

def extract_items(driver):
    """현재 페이지 아이템 추출 함수"""
    items = driver.find_elements(By.CSS_SELECTOR, "div.itemUnit")
    results = []
    for item in items:
        try:
            title = item.find_element(By.CSS_SELECTOR, "a.gd_name").text.strip()
        except:
            title = "없음"
        try:
            author_tags = item.find_elements(By.CSS_SELECTOR, "span.authPub.info_auth a")
            author = ", ".join([a.text.strip() for a in author_tags if a.text.strip()])
        except:
            author = "없음"
        try:
            rating = item.find_element(By.CSS_SELECTOR, "span.rating_grade em.yes_b").text.strip()
        except:
            rating = "없음"
        try:
            price = item.find_element(By.CSS_SELECTOR, "strong.txt_num em.yes_b").text.strip()
        except:
            price = "없음"
        try:
            salesnum = item.find_element(By.CSS_SELECTOR, "span.saleNum").text.strip()
        except:
            salesnum = "없음"
        try:
            rating_rvCount = item.find_element(By.CSS_SELECTOR, "span.rating_rvCount").text.strip()
        except:
            rating_rvCount = "없음"

        results.append((title, author, rating, price, salesnum, rating_rvCount))
    return results

# 2. 1페이지 수집
all_items = extract_items(driver)
print(f"1페이지 {len(all_items)}개 수집 완료")

# 3. 2페이지 버튼 클릭 (title="2" 인 a.num 태그)
try:
    page2_btn = driver.find_element(By.CSS_SELECTOR, "a.num[title='2']")
    page2_btn.click()
    time.sleep(3)
    page2_items = extract_items(driver)
    all_items.extend(page2_items)
    print(f"2페이지 {len(page2_items)}개 수집 완료")
except:
    print("2페이지 버튼을 찾지 못했습니다.")

# 4. 출력
print(f"\n총 {len(all_items)}개 수집\n" + "="*60)

print()
print("\n-------------------------------------------------------------------")
print("----------------------⭐⭐이번 주 베스트⭐⭐----------------------")
print("-------------------------------------------------------------------\n")

for i, (title, author, rating, price, salesnum, rating_rvCount) in enumerate(all_items, 1):
    print(f"[{i:2}위] 제목: {title} | 저자: {author} | 평점: {rating}")
    print(f"       가격: {price}원 | {salesnum} | 리뷰: {rating_rvCount}\n")

driver.quit()
```
