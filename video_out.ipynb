```python
import cv2
cap = cv2.VideoCapture('video_out.mp4')


while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imshow('video',frame)
    if cv2.waitKey(20) == ord('q'):
        break 

cap.release()
cv2.destroyAllWindows()

```

```python
cap = cv2.VideoCapture('video_out.mp4')

#코덱정의하는 부분
# 영상 파일은 용량이 매우 커서 압축해서 저장하고 풀어서 재생해야한다
fourcc = cv2.VideoWriter_fourcc(*'DIVX') #한글자씩 띄어서 입력 *
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # 영상 크기는 얼마로 할지
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # 영상 크기는 얼마로 할지
fps = cap.get(cv2.CAP_PROP_FPS) # 몇 fps로 저장할지


out= cv2.VideoWriter('video_out.avi',fourcc,fps,(width,height)) # 어떤 이름으로 저장할지



while cap.isOpened():  #영상파일이 열려있는 동안 계속해서 반복한다.
    ret, frame = cap.read()  #프레임을 읽는다.  
    #ret >> 읽기 성공 여부 T/F 
    #frame >> 읽어온 이미지 데이터
    
    if not ret: #프레임이 더없으면 break(더 재생될 영상이 없으면) 
        break

    #읽어온 이미지데이터를 시계반대방향으로 90도 만큼 회전시킨다
    frame = cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE)
    
                       
    out.write(frame)
    cv2.imshow('video',frame)

    # waitKey(이미지와 이미지사이의 간격) 수치를 줄여서 동영상의 속도를 높힌다
    if cv2.waitKey(5) == ord('q'):
        break

out.release()
cap.release()
cv2.destroyAllWindows()
```
