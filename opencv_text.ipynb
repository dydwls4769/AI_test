### 텍스트
- 글꼴 OpenCV에서 사용하는 글꼴 종류
    - cv2.FONT_HERSHEY_SIMPLEX : 보통 크기의 산 세리프(sans-serif) 글꼴
    - cv2.FONT_HERSHEY_PLAIN : 작은 크기의 산 세리프 글꼴
    - cv2.FONT_HERSHEY_SCRIPT_SIMPLEX : 필기체 스타일 글꼴
    - cv2.FONT_HERSHEY_TRIPLEX : 보통 크기의 세리프 글꼴
    - cv2.FONT_ITALIC : 기울임

### cv2.putText(그릴위치, 텍스트내용, 시작위치, 폰트종류, 크기, 색깔, 두께) 글자를 화면에 표시하기 위함

```python
import cv2
import numpy as np

img = np.zeros((480,640,3), dtype=np.uint8)

SCALE = 2
COLOR = (255,255,255)
THIC = 2

cv2.putText(img, "Korea Fighting!!",(20,50),cv2.FONT_HERSHEY_SIMPLEX, SCALE, COLOR, THIC)
cv2.putText(img, "Korea Fighting!!",(20,150),cv2.FONT_HERSHEY_PLAIN, SCALE, COLOR, THIC)
cv2.putText(img, "Korea Fighting!!",(20,250),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, SCALE, COLOR, THIC)
cv2.putText(img, "Korea Fighting!!",(20,350),cv2.FONT_HERSHEY_TRIPLEX, SCALE, COLOR, THIC)
cv2.putText(img, "Korea Fighting!!",(20,450),cv2.FONT_HERSHEY_TRIPLEX | cv2.FONT_ITALIC, SCALE, COLOR, THIC) #기울임꼴로

cv2.imshow('text', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

-한글 우회 출력 

```python
import cv2
import numpy as np
#PIL(python image library) >> 글자를 직접 그려버린다
from PIL import ImageFont, ImageDraw, Image

def myputText(src, text, pos, font_size, font_color):
    img_pil = Image.fromarray(src)
    draw = ImageDraw.Draw(img_pil)
    font = ImageFont.truetype('fonts/gulim.ttc', font_size)
    draw.text(pos, text, font=font, fill = font_color)
    return np.array(img_pil)
    

img = np.zeros((480,640,3), dtype=np.uint8)

FONT_SIZE = 30
COLOR = (255,255,255)
#THIC = 1

#cv2.putText(img, "대한민국 파이팅!!",(20,50),cv2.FONT_HERSHEY_SIMPLEX, SCALE, COLOR, THIC) #오픈CV는 기본적으로 한글지원X
img = myputText(img, "대한민국 파이팅!!", (20,50), FONT_SIZE, COLOR)

cv2.imshow('text', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture('video.mp4')

##코덱 정의하는 부분
fourcc = cv2.VideoWriter_fourcc(*'DIVX') #한글자씩 띄어서 입력 *
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

out= cv2.VideoWriter('output.avi',fourcc,fps,(width,height))



while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    out.write(frame)
    
    cv2.imshow('video',frame)
    if cv2.waitKey(20) == ord('q'):
        break 

out.release()
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture('video.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_resize = cv2.resize(frame, (400,500))
    cv2.imshow('resize',frame_resize)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture('video.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_resize = cv2.resize(frame,None, fx=5, fy=2, interpolation=cv2.INTER_CUBIC)
    cv2.imshow('resize',frame_resize)
    if cv2.waitKey(5) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
