```python
#라이브러리 임포트
import requests
import lxml
from bs4 import BeautifulSoup
```

```python
#타겟사이트 url
url = "http://quotes.toscrape.com/"
```

```python
#웹사이트에 접속하고 요청
response = requests.get(url)
```

```python
#웹사이트에 접속요청을 보냄 200=성공 404=페이지 없음 403=접근거부
if response.status_code == 200: 
    html= response.text #response안에서 html텍스트만 꺼내서 'html'변수에 저장
    soup= BeautifulSoup(html, 'lxml') #문자열인 html을 파이썬이 탐색 가능한 객체로 변환 lxml은 피싱방식

    quotes = soup.select('.quote')
    print(f'총 {len(quotes)}개의 명언을 찾았습니다.\n')

    #데이터를 추출, 출력
    for i , q in enumerate(quotes, 1):
        #quotes 리스트를 하나씩 꺼내면서 반복 enumerate는 숫자번호도 같이 줌
        #명언 텍스트 추출
        text = q.select_one('.text').text
        #현재 명언카드('q')안에서 class=text 태그 1개찾기
        #저자 이름 추출
        author= q.select_one('.author').text
        #현재 명언카드('q')안에서 class=author 태그 찾고 글자 추출

        print(f'{i}. {text} ')
        print(f' -저자 : {author}')
        print('='*50)
        
else:
    print("웹 사이트 접속에 실패",response.status_code)
```

```python

base_url = "https://quotes.toscrape.com/page/{}/" # /{}/ 페이지 수를 가져올수있게 
all_quotes = [] #리스트를 만들어서 저장

for page in range(1, 6): #1~6
    response = requests.get(base_url.format(page))
    
    if response.status_code == 200: #200이면 정상 
        html  = response.text
        soup  = BeautifulSoup(html, 'lxml')
        quotes = soup.select('.quote')
        
        for i, q in enumerate(quotes, 1):  
            text   = q.select_one('.text').text
            author = q.select_one('.author').text
            tags   = [t.text for t in q.select('.tag')]
            all_quotes.append({'text': text, 'author': author, 'tags': tags}) 


for i, q in enumerate(all_quotes, 1):
    print(f'{i}. {q["text"]}')
    print(f'   - 저자: {q["author"]}')
    print(f'   - 태그: {q["tags"]}')
    print('='*50)
```
